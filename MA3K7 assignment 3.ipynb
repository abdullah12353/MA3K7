```python
#defining a random hat game with n numbers
#def hatgame(n, iterations):
import random 
import matplotlib.pyplot as plt
import numpy as np

#simulating a random hat game with n numbers
def random_hat_game(n):
    numbers = list(range(1, n+1))
    counter = n
    while counter > 1:
        nums = random.sample(range(counter), 2)
        if numbers[nums[0]] <= numbers[nums[1]]:
            difference = numbers[nums[1]] - numbers[nums[0]]
            numbers[nums[0]] = difference
            numbers.pop(nums[1])
    
        else:
            difference = numbers[nums[0]] - numbers[nums[1]]
            numbers[nums[1]] = difference
            numbers.pop(nums[0])
        counter -= 1
    return(numbers[0])

#plotting histogram of results of simulation for a given n
def histogramplot(n, iter):
    data = np.zeros((iter,))
    for i in range(iter):
        data[i] = random_hat_game(n)
    bins = np.arange(0, data.max() + 1.5) - 0.5
    fig, ax = plt.subplots()
    _ = ax.hist(data, bins)
    ax.set_xticks(bins + 0.5)
    plt.title("histogram plot for hat game with "+str(n)+" numbers")

    
#checking if game has any odd results
def oddcheck(n, iter):
    data = np.zeros((iter,))
    for i in range(iter):
        data[i] = random_hat_game(n)
    odds = data[data % 2 != 0]
    print("odds results are: ")
    print(odds)
    
#producing plots and results
histogramplot(3,10000)
histogramplot(4,10000)
histogramplot(5,10000)
histogramplot(6,10000)
histogramplot(7,10000)
histogramplot(8,10000)
oddcheck(2024,10000)
```
