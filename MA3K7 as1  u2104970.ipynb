```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
def find_rep_period(num1, num2):
    seq = [num1, num2]
    
    #using dictionary to check for repeats
    prevseen_patterns = {(num1, num2): 0}

    # using a for loop to prevent crash from inifinite loop
    for j in range(1, 1000):
        
        #calculating next term in sequence        
        next_num = (seq[-1]+seq[-2]) % 10
        seq.append(next_num)

        current_pattern = (seq[-2], seq[-1])
    
    
        #checking if the newly calculated values are a repeat of any previous ones
        if current_pattern in prevseen_patterns:
            return j

        prevseen_patterns[current_pattern] = j

    return None  # If no repetitions are found

periods=[]
count  = 0

per = np.empty((100,3))
for a in range(10):
    for b in range(10):
        result_period = find_rep_period(a,b)
        periods.append(result_period)
        per[count][0] = a
        per[count][1] = b
        per[count][2] = result_period
        count+=1
        
print(periods)
print(per)
```
