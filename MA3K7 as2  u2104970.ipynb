```python
import numpy as np
import matplotlib.pyplot as plt

def prob_of_winning(n, starting):
    turn_splitter = int(n**2/2)
    if n%2 == 1:
        #if we are starting
        if starting == True: 
            M = np.array((1,)*turn_splitter+(0,)*(turn_splitter+1))
        else:
            M = np.array((1,)*(turn_splitter+1)+(0,)*turn_splitter)
    else:
        M = np.array((1,)*turn_splitter+(0,)*turn_splitter)
    wins = 0
    #producing 20000 sample games
    for i in range(20000):
        np.random.shuffle(M)
        det = np.linalg.det(np.reshape(M,(n,n))) 
        if det == 0:
            wins+=1
    #returns proportion of games won in random scenario
    return wins/20000



dimensions = np.arange(2,20)
startfirst= []
startsecond = [] 
for i in dimensions:
    startfirst.append(prob_of_winning(i,True))
    startsecond.append(prob_of_winning(i,False))
    
plt.scatter(dimensions,startfirst, label="If I start")
plt.scatter(dimensions,startsecond, label="If my opponent starts")
plt.grid()
plt.xlabel('Dimension')
plt.ylabel('Winning Probability')
plt.legend()
plt.show()
```
