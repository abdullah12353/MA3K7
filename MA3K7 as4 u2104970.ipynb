```python
#want to simulate random walk, step size either 1 or 2
#end simulation after 24 steps
import random
import matplotlib.pyplot as plt
import numpy as np

#defining a function that counts how many times you hit square n, with iters number of random walks
def random_walk(n,iters):
    hits = 0
    for i in range(iters):
        counter = 0
        pos = 1
        #a single random walk, which has at most n moves
        while counter < n+1:
            step_size =  random.choice([1, 2])
            pos += step_size
            if pos == n:
                hits += 1
                break
            if pos == (n+1):
                break
    return hits

#defining parameters
iters=100000
n=25

#calculating the number of hits
z=random_walk(n,iters)
labels = ['Hit 25', 'Missed 25']
sizes = [z, iters-z]

#plotting number of hits vs misses in pie chart
fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.show()


#plotting our h(n) vs n

#generating array of integers 1-20
t = np.arange(1,21)

#defining h(n) as we found
def diffeqn(x):
    return (2/3)+(-2/3)*(-1/2)**(x)

#applying h(n) to our array of integers
vfunc = np.vectorize(diffeqn)
result = vfunc(t)

#generating the plot
fig, ax = plt.subplots()
ax.plot(t, result)
ax.hlines(y=2/3, xmin=0, xmax=20, color='r')
plt.legend(["probabilites", "p=2/3"])
plt.xlabel("square to land on")
plt.ylabel("hitting probability")
plt.show()
```
